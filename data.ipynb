```python
with open('movies.txt', 'w') as file:
    # Iterate over each row in the DataFrame
    for index, row in df.iterrows():
        # Write the title and data for each column
        file.write(f"{row['movie_title']}:\n")
        file.write(
            f"Title of the Movie: {row['movie_title']}\n")
        file.write(f"Duration in minutes: {row['duration']}\n")
        file.write(
            f"Name of the Director of the Movie : {row['director_name']}\n")
        file.write(
            f"Number of likes of the Director on his Facebook Page : {row['director_facebook_likes']}\n")
        file.write(
            f"Primary actor starring in the movie: {row['actor_1_name']}\n")
        file.write(
            f"Number of likes of the Actor_1 on his/her Facebook Page: {row['actor_1_facebook_likes']}\n")
        file.write(
            f"Other actor starring in the movie: {row['actor_2_name']}\n")
        file.write(
            f"Number of likes of the Actor_2 on his/her Facebook Page: {row['actor_2_facebook_likes']}\n")
        file.write(
            f"Other actor starring in the movie: {row['actor_3_name']}\n")
        file.write(
            f"Number of likes of the Actor_3 on his/her Facebook Page: {row['actor_3_facebook_likes']}\n")
        file.write(
            f"Number of users who gave a review: {row['num_user_for_reviews']}\n")
        file.write(
            f"Number of critical reviews on imdb: {row['num_critic_for_reviews']}\n")
        file.write(
            f"Number of people who voted for the movie: {row['num_voted_users']}\n")
        file.write(
            f"Total number of facebook likes of the entire cast of the movie: {row['cast_total_facebook_likes']}\n")
        file.write(
            f"Number of Facebook likes in the movie page: {row['movie_facebook_likes']}\n")
        file.write(
            f"Keywords describing the movie plot: {row['plot_keywords']}\n")
        file.write(
            f"Number of the actor who featured in the movie poster: {row['facenumber_in_poster']}\n")
        file.write(
            f"Film colorization. 'Black and White' or 'Color': {row['color']}\n")
        file.write(f"Film categorization: {row['genres']}\n")
        file.write(
            f"The year in which the movie is released: {row['title_year']}\n")
        file.write(f"Movie Language: {row['language']}\n")
        file.write(
            f"Country where the movie is produced: {row['country']}\n")
        file.write(f"Content rating of the movie: {row['content_rating']}\n")
        file.write(
            f"Aspect ratio the movie was made in: {row['aspect_ratio']}\n")
        file.write(f"IMDB link of the movie: {row['movie_imdb_link']}\n")
        file.write(f"Gross earnings of the movie in Dollars: {row['gross']}\n")
        file.write(f"Budget of the movie in Dollars: {row['budget']}\n")
        file.write(f"IMDB Score of the movie on IMDB: {row['imdb_score']}\n")
        if row['imdb_score'] >= 1 and row['imdb_score']<3:
            file.write(f"Movie Status: Flop Movie\n")
        elif row['imdb_score'] >= 3 and row['imdb_score'] < 6:
            file.write(f"Movie Status: Average Movie\n")
        elif row['imdb_score'] >= 6 and row['imdb_score'] <= 10:
            file.write(f"Movie Status: Hit Movie\n")

        file.write("\n")
        file.write("\n")
        file.write("\n")
```

```python
with open('hitmovies.txt', 'w') as file:
    # Iterate over each row in the DataFrame
    file.write("Hit Movies\n\n\n")

    for index, row in df.iterrows():
        # Write the title and data for each column
        if row['imdb_score'] >= 6 and row['imdb_score'] <= 10:
            file.write(f"{row['movie_title']}\n")
```

```python
with open('averagemovies.txt', 'w') as file:
    # Iterate over each row in the DataFrame
    file.write("Average Movies\n\n\n")

    for index, row in df.iterrows():
        # Write the title and data for each column
        if row['imdb_score'] >= 3  and row['imdb_score'] < 6:
            file.write(f"{row['movie_title']}\n")
```

```python
with open('flopmovies.txt', 'w') as file:
    # Iterate over each row in the DataFrame
    file.write("Flop Movies\n\n\n")

    for index, row in df.iterrows():
        # Write the title and data for each column
        if row['imdb_score'] >= 0 and row['imdb_score'] < 3:
            file.write(f"{row['movie_title']}\n")
```
